```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1Ô∏è‚É£ Load Dataset
df = pd.read_csv("C:/Users/jsrv7/.vscode/cleaned_dataset.csv")
print("‚úÖ Dataset Loaded Successfully!\n")

# 2Ô∏è‚É£ Identify Columns
print(df.info(), "\n")

# Try to detect region or location column automatically
region_col = None
for col in df.columns:
    if 'region' in col.lower() or 'city' in col.lower() or 'location' in col.lower() or 'area' in col.lower():
        region_col = col
        break

if not region_col:
    region_col = input("Enter the column name representing regions (e.g., Region, City, Location): ")

print(f"üìç Using region column: {region_col}\n")

# Detect numeric weather columns (e.g., temperature, rainfall, humidity)
numeric_cols = df.select_dtypes(include='number').columns.tolist()

if not numeric_cols:
    print("‚ö† No numeric columns found for comparison.")
    exit()

print(f"üå° Numeric weather columns detected: {numeric_cols}\n")

# 3Ô∏è‚É£ Statistical Summary by Region
region_summary = df.groupby(region_col)[numeric_cols].mean().round(2)
print("üìä Average Weather Conditions by Region:\n")
print(region_summary, "\n")

# 4Ô∏è‚É£ Visual Comparison (Bar Charts)
for col in numeric_cols:
    plt.figure(figsize=(8, 5))
    sns.barplot(x=region_col, y=col, data=df, palette="viridis")
    plt.title(f"Average {col} by {region_col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# 5Ô∏è‚É£ Correlation Comparison
plt.figure(figsize=(8, 6))
sns.heatmap(region_summary.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Weather Parameters (Regional Averages)")
plt.tight_layout()
plt.show()

# 6Ô∏è‚É£ Save Summary
region_summary.to_csv("region_weather_comparison.csv")
print("üíæ Regional weather comparison saved to 'region_weather_comparison.csv'")
print("‚úÖ Comparison Completed!")



```
