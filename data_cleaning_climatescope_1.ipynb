#### Cleaning 'Country' column

```python
def clean_countries(df, country_col='country'):
    
    
    
    # Ensure the column exists
    if country_col not in df.columns:
        raise ValueError(f"Column '{country_col}' not found in DataFrame.")
    
    # Convert to string
    df[country_col] = df[country_col].astype('string')

    # Dictionary for corrections (typos or language variations)
    corrections = {
        "Kyrghyzstan": "Kyrgyzstan",
        "Komoren": "Comoros",
        "Inde": "India",
        "Malásia": "Malaysia",
        "Estonie": "Estonia",
        "Letonia": "Latvia",
        "Mexique": "Mexico",
        "Marrocos": "Morocco",
        "Saudi Arabien": "Saudi Arabia",
        "Südkorea": "South Korea",
        "Turkménistan": "Turkmenistan",
        "Jemen": "Yemen",
        "USA United States of America": "United States of America",
        "Saint-Vincent-et-les-Grenadines": "Saint Vincent and the Grenadines",
        "Bélgica": "Belgium",
        "Polônia": "Poland",
        "Lao People's Democratic Republic": "Laos",
        "Cote d'Ivoire": "Ivory Coast"
    }

    # Replace wrong or non-standard names
    df[country_col] = df[country_col].replace(corrections)

    # Non-Latin name replacements
    non_latin = {
        "كولومبيا": "Colombia",
        "Гватемала": "Guatemala",
        "Польша": "Poland",
        "Турция": "Turkey",
        "火鸡": "Turkey"
    }
    df[country_col] = df[country_col].replace(non_latin)

    # Strip whitespace and coverts all to lower cases
    df[country_col] = df[country_col].str.strip().str.lower()

    return df
print(clean_countries(dataset, country_col='country')['country'].unique().tolist())
```

#### Aggregate weather


```python
import pandas as pd

def get_season(month):
    
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Autumn"

def aggregate_weather(df, date_col='last_updated', num_cols=None, save_csv=True):
    
    # Ensure datetime
    df[date_col] = pd.to_datetime(df[date_col], errors='coerce',dayfirst=True)
    
    # Default numeric columns if not provided
    if num_cols is None:
        num_cols = [
            'temperature_celsius', 'wind_kph', 'wind_degree', 'pressure_mb', 'precip_mm',
            'humidity', 'cloud', 'feels_like_celsius', 'visibility_km', 'uv_index',
            'gust_kph', 'air_quality_Carbon_Monoxide', 'air_quality_Ozone',
            'air_quality_Nitrogen_dioxide', 'air_quality_Sulphur_dioxide',
            'air_quality_PM2.5', 'air_quality_PM10'
        ]
    
    # MONTHLY aggregation
    df['year_month'] = df[date_col].dt.to_period('M')
    monthly_avg = df.groupby(['country', 'location_name', 'year_month'])[num_cols].mean(numeric_only=True).reset_index()
    
    # YEARLY aggregation
    df['year'] = df[date_col].dt.year
    yearly_avg = df.groupby(['country', 'location_name', 'year'])[num_cols].mean(numeric_only=True).reset_index()
    
    # SEASONAL aggregation
    df['month'] = df[date_col].dt.month
    df['season'] = df['month'].apply(get_season)
    seasonal_avg = df.groupby(['country', 'location_name', 'year', 'season'])[num_cols].mean(numeric_only=True).reset_index()
    
    # Save CSV files if requested
    if save_csv:
        monthly_avg.to_csv("weather_data_monthly.csv", index=False)
        yearly_avg.to_csv("weather_yearly_avg.csv", index=False)
        seasonal_avg.to_csv("weather_seasonal_avg.csv", index=False)
    
    return {
        'monthly_avg': monthly_avg,
        'yearly_avg': yearly_avg,
        'seasonal_avg': seasonal_avg
    }
aggregated = aggregate_weather(dataset)




```
