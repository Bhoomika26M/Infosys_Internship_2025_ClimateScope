***Milestone 2: Core Analysis & Visualization Design***

```python
numeric_cols = [
    "temperature_celsius", "feels_like_celsius", "humidity", "wind_kph",
    "gust_kph", "pressure_mb", "precip_mm", "cloud", "visibility_km", "uv_index",
    "air_quality_Carbon_Monoxide", "air_quality_Ozone", "air_quality_Nitrogen_dioxide",
    "air_quality_Sulphur_dioxide", "air_quality_PM2.5", "air_quality_PM10",
    "air_quality_us-epa-index", "air_quality_gb-defra-index"
]
```

```python
print("Basic Statistics:")
display(df_clean[numeric_cols].describe())
```

```python
from scipy.stats import skew, kurtosis

print("Skewness & Kurtosis:\n")
for col in numeric_cols:
    col_skew = skew(df_clean[col].dropna())
    col_kurt = kurtosis(df_clean[col].dropna())
    print(f"{col:<35} Skew = {col_skew:>7.2f}    Kurtosis = {col_kurt:>7.2f}")
```

```python
# Collect values
stats = []
for col in numeric_cols:
    stats.append([col, skew(df_clean[col].dropna()), kurtosis(df_clean[col].dropna())])

df_stats = pd.DataFrame(stats, columns=["Variable", "Skewness", "Kurtosis"])

# Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(
    df_stats.set_index("Variable"),
    annot=True,
    cmap="RdYlGn",      # Red → Yellow → Green gradient
    center=0,
    fmt=".1f"
)
plt.title("Skewness & Kurtosis of Numeric Features", fontsize=14, weight="bold")
plt.show()

```

```python
from scipy.stats import zscore
import numpy as np

# Make a copy of the dataset
df_cleaned = df_clean.copy()

# Select columns to clean
cols_to_clean = ["pressure_mb", "air_quality_Sulphur_dioxide"]

for col in cols_to_clean:
    col_data = df_cleaned[col].dropna()
    z_scores = zscore(col_data)

    # Mask: True for values within ±3 std deviations
    mask = abs(z_scores) < 3

    # Replace outliers with NaN (only within valid indices)
    df_cleaned.loc[col_data.index, col] = col_data.where(mask, np.nan)

# Compare before vs after
print("Before cleaning:")
print(df_clean[cols_to_clean].describe())

print("\nAfter cleaning:")
print(df_cleaned[cols_to_clean].describe())
```

```python
from scipy.stats import skew, kurtosis

print("\nSkewness & Kurtosis after cleaning:\n")
for col in numeric_cols:
    # Check if column exists in df_cleaned (sometimes numeric_cols list may include dropped columns)
    if col in df_cleaned.columns:
        col_data = df_cleaned[col].dropna()
        if len(col_data) > 0:  # avoid empty-column errors
            col_skew = skew(col_data)
            col_kurt = kurtosis(col_data)
            print(f"{col:<35} Skew = {col_skew:>7.2f}   Kurtosis = {col_kurt:>7.2f}")
        else:
            print(f"{col:<35} Skew =  N/A     Kurtosis =  N/A (no data)")
    else:
        print(f"{col:<35} Not found in df_cleaned")
```

```python
# Collect values
stats = []
for col in numeric_cols:
    stats.append([col, skew(df_cleaned [col].dropna()), kurtosis(df_cleaned [col].dropna())])

df_stats = pd.DataFrame(stats, columns=["Variable", "Skewness", "Kurtosis"])

# Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(
    df_stats.set_index("Variable"),
    annot=True,
    cmap="RdYlGn",      # Red → Yellow → Green gradient
    center=0,
    fmt=".1f"
)
plt.title("Skewness & Kurtosis of Numeric Features", fontsize=14, weight="bold")
plt.show()

```

```python
# Extract month from datetime
df_cleaned["month"] = df_cleaned["last_updated"].dt.month

# Add more important seasonal columns
cols_to_analyze = [
    "temperature_celsius",
    "precip_mm",
    "humidity",
    "wind_kph",
    "air_quality_PM2.5",
    "air_quality_PM10",
    "uv_index"
]
# Compute monthly averages
monthly_avg = df_cleaned.groupby("month")[cols_to_analyze].mean()
# Plot
monthly_avg.plot(kind="line", marker="o", figsize=(12, 7))
plt.title("Seasonal Patterns: Weather & Air Quality Indicators")
plt.xlabel("Month")
plt.ylabel("Average Value")
plt.legend(title="Indicators")
plt.grid(True)
plt.show()
```

```python
import plotly.express as px

# Aggregate monthly average temperature per country
monthly_temp = df_cleaned.groupby([df_cleaned['country'], df_cleaned['last_updated'].dt.month])['temperature_celsius'].mean().reset_index()
monthly_temp.rename(columns={'last_updated': 'month'}, inplace=True)

# Plot choropleth with hot-red / cold-blue theme
fig = px.choropleth(
    monthly_temp,
    locations="country",
    locationmode="country names",
    color="temperature_celsius",
    hover_name="country",
    animation_frame="month",
    color_continuous_scale="RdBu_r",   # red = hot, blue = cold
    title="Monthly Average Temperature by Country"
)
fig.show()
```

```python
#  Define thresholds for "extreme" values
extreme_conditions = {
    "heatwave": df_cleaned["temperature_celsius"] > 40,   # >40°C
    "cold_wave": df_cleaned["temperature_celsius"] < 0,   # <0°C
    "heavy_rain": df_cleaned["precip_mm"] > 100,          # >100 mm rainfall
    "storm": df_cleaned["wind_kph"] > 80,                 # strong winds
    "high_pollution": df_cleaned["air_quality_PM2.5"] > 150  # unhealthy air
}

#  Create a dictionary of extreme events
extreme_events = {}
for event, condition in extreme_conditions.items():
    extreme_events[event] = df_cleaned[condition][
        ["country", "location_name", "last_updated",
         "temperature_celsius", "precip_mm", "wind_kph", "air_quality_PM2.5"]
    ]

# Example: Show first 5 heatwave events
print(" Heatwaves Detected:")
display(extreme_events["heatwave"].head())
```

```python
# Count number of extreme events per month
extreme_counts = {}
for event, df_event in extreme_events.items():
    if not df_event.empty:
        df_event = df_event.copy()
        df_event["month"] = df_event["last_updated"].dt.month
        extreme_counts[event] = df_event.groupby("month").size()

# Plot each extreme event trend
plt.figure(figsize=(12, 6))
colors = ["red", "orange", "blue", "green", "purple"]  # customize per event

for (event, counts), color in zip(extreme_counts.items(), colors):
    plt.plot(counts.index, counts.values, marker="o", label=event, color=color)

plt.legend(title="Extreme Event Type", fontsize=10)
plt.title("Monthly Trends of Extreme Weather Events", fontsize=14, weight="bold")
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Events", fontsize=12)
plt.xticks(range(1, 13))  # show months 1 to 12
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
top_countries = pd.concat(extreme_events.values())["country"].value_counts().head(10)

plt.figure(figsize=(16,8))
sns.barplot(
    x=top_countries.index,
    y=top_countries.values,
    hue=top_countries.index,  # assign x to hue
    palette="viridis",
    dodge=False,
    legend=False
)
plt.xticks(rotation=45, ha="right")
plt.title("Top 10 Countries with Most Extreme Weather Events", fontsize=18, weight="bold")
plt.ylabel("Number of Events", fontsize=14)
plt.xlabel("Country", fontsize=14)
plt.tight_layout()
plt.show()
```
