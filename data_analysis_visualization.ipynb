#### Counting Outliers

```python
import numpy as np

def count_outliers(df):
    summary = {}

    # Air Quality columns
    air_quality_cols = [c for c in df.columns if 'air_quality' in c]
    for col in air_quality_cols:
        outliers = df[(df[col] < 0) | (df[col] > 5000)][col].count()
        summary[col] = outliers

    # Wind speed
    summary['wind_kph'] = df[df['wind_kph'] > 150]['wind_kph'].count()

    # Pressure
    summary['pressure_mb'] = df[(df['pressure_mb'] < 900) | (df['pressure_mb'] > 1100)]['pressure_mb'].count()

    # Temperature
    for col in ['temperature_celsius', 'feels_like_celsius']:
        if col in df.columns:
            summary[col] = df[(df[col] < -50) | (df[col] > 60)][col].count()

    return summary
outlier_counts = count_outliers(daily)    #outliers count for daily data   ( more count so replace with NaN values later handle them)
#outlier_counts = count_outliers(monthly) #outliers count for monthly data (less count in monthly and yearly so we can ignore them)
#outlier_counts = count_outliers(yearly)  ##outliers count for yearly data
pd.DataFrame.from_dict(outlier_counts, orient='index', columns=['Outlier Count'])
```

#### Handling outliers

```python
import numpy as np
import pandas as pd

def clean_and_fill_daily(df, apply_rolling=False):
    
    
    summary = {}
    
    # Identify air quality columns
    air_quality_cols = [c for c in df.columns if 'air_quality' in c]
    
    # ---- Step 1: Replace outliers with NaN ----
    # Air Quality
    for col in air_quality_cols:
        outliers = df[(df[col] < 0) | (df[col] > 5000)][col].count()
        summary[col] = outliers
        df[col] = df[col].apply(lambda x: np.nan if x < 0 or x > 5000 else x)
    
    # Wind speed
    summary['wind_kph'] = df[df['wind_kph'] > 150]['wind_kph'].count()
    df['wind_kph'] = df['wind_kph'].apply(lambda x: np.nan if x > 150 else x)
    
    # Pressure
    summary['pressure_mb'] = df[(df['pressure_mb'] < 900) | (df['pressure_mb'] > 1100)]['pressure_mb'].count()
    df['pressure_mb'] = df['pressure_mb'].apply(lambda x: np.nan if x < 900 or x > 1100 else x)
    
    # Temperature
    for col in ['temperature_celsius', 'feels_like_celsius']:
        if col in df.columns:
            summary[col] = df[(df[col] < -50) | (df[col] > 60)][col].count()
            df[col] = df[col].apply(lambda x: np.nan if x < -50 or x > 60 else x)
    
    # ---- Step 2: Fill NaN values ----
    # Linear interpolation (works for RangeIndex)
    df[air_quality_cols] = df[air_quality_cols].interpolate(method='linear')
    
    # Optional: apply rolling median to smooth data
    if apply_rolling:
        df[air_quality_cols] = df[air_quality_cols].rolling(window=7, center=True, min_periods=1).median()
    
    # Display summary
    print("\n Outlier Replacement Summary:")
    display(pd.DataFrame.from_dict(summary, orient='index', columns=['Outlier Count']))
    print("\nNaN values have been filled using interpolation and optional rolling median.\n")
    
    return df
daily_cleaned = clean_and_fill_daily(daily, apply_rolling=True)
```
